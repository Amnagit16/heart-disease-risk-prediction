```python
# Step 1: Import required libraries

import pandas as pd
import numpy as np
import os

pd.set_option("display.max_columns", None)
```

```python
# Step 2: Load prepared dataset

df = pd.read_csv("../Data/heart_disease_prepared.csv")

df.head()
```

```python
# Step 3: Check dataset shape

df.shape
```

```python
# Step 4: Create Tableau dashboard data folder

os.makedirs("../tableau_dashboard_data", exist_ok=True)
```

```python
# Step 5: Confirm Tableau folder was created

os.listdir("..")
```

```python
# Step 6: Create target distribution summary for Tableau

target_summary = df["HeartDiseaseorAttack"].value_counts().reset_index()
target_summary.columns = ["HeartDiseaseorAttack", "Count"]

target_summary["Heart Disease Status"] = target_summary["HeartDiseaseorAttack"].map({
    0.0: "No Heart Disease / Attack",
    1.0: "Heart Disease / Attack"
})

target_summary["Percentage"] = (target_summary["Count"] / target_summary["Count"].sum()) * 100

target_summary
```

```python
# Step 7: Save target distribution summary

target_summary.to_csv("../tableau_dashboard_data/target_distribution_summary.csv", index=False)
```

```python
# Step 8: Confirm saved Tableau file

os.listdir("../tableau_dashboard_data")
```

```python
# Step 9: Create heart disease rate by age group

age_summary = df.groupby("Age")["HeartDiseaseorAttack"].agg(
    Count="count",
    Heart_Disease_Rate="mean"
).reset_index()

age_summary["Heart_Disease_Rate"] = age_summary["Heart_Disease_Rate"] * 100

age_summary["Age Group"] = age_summary["Age"].map({
    1.0: "18-24",
    2.0: "25-29",
    3.0: "30-34",
    4.0: "35-39",
    5.0: "40-44",
    6.0: "45-49",
    7.0: "50-54",
    8.0: "55-59",
    9.0: "60-64",
    10.0: "65-69",
    11.0: "70-74",
    12.0: "75-79",
    13.0: "80+"
})

age_summary
```

```python
# Step 10: Save age group summary

age_summary.to_csv("../tableau_dashboard_data/age_group_heart_disease_rate.csv", index=False)
```

```python
# Step 11: Confirm saved Tableau files

os.listdir("../tableau_dashboard_data")
```

```python
# Step 12: Create heart disease rate by key risk factors

risk_factors = [
    "HighBP",
    "HighChol",
    "Diabetes",
    "Smoker",
    "Stroke",
    "PhysActivity",
    "DiffWalk"
]

risk_factor_summary_list = []

for feature in risk_factors:
    temp = df.groupby(feature)["HeartDiseaseorAttack"].agg(
        Count="count",
        Heart_Disease_Rate="mean"
    ).reset_index()
    
    temp["Heart_Disease_Rate"] = temp["Heart_Disease_Rate"] * 100
    temp["Feature"] = feature
    temp = temp.rename(columns={feature: "Feature_Value"})
    
    risk_factor_summary_list.append(temp)

risk_factor_summary = pd.concat(risk_factor_summary_list, ignore_index=True)

risk_factor_summary
```

```python
# Step 13: Save risk factor summary

risk_factor_summary.to_csv("../tableau_dashboard_data/risk_factor_heart_disease_rate.csv", index=False)
```

```python
# Step 14: Confirm saved Tableau files

os.listdir("../tableau_dashboard_data")
```

```python
# Step 15: Create heart disease rate by general health

general_health_summary = df.groupby("GenHlth")["HeartDiseaseorAttack"].agg(
    Count="count",
    Heart_Disease_Rate="mean"
).reset_index()

general_health_summary["Heart_Disease_Rate"] = general_health_summary["Heart_Disease_Rate"] * 100

general_health_summary["General Health"] = general_health_summary["GenHlth"].map({
    1.0: "Excellent",
    2.0: "Very Good",
    3.0: "Good",
    4.0: "Fair",
    5.0: "Poor"
})

general_health_summary
```

```python
# Step 16: Save general health summary

general_health_summary.to_csv("../tableau_dashboard_data/general_health_heart_disease_rate.csv", index=False)
```

```python
# Step 17: Confirm saved Tableau files

os.listdir("../tableau_dashboard_data")
```

```python
# Step 18: Create BMI category column for Tableau summary

def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal weight"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["BMI_Category"] = df["BMI"].apply(bmi_category)

df[["BMI", "BMI_Category"]].head()
```

```python
# Step 19: Create heart disease rate by BMI category

bmi_summary = df.groupby("BMI_Category")["HeartDiseaseorAttack"].agg(
    Count="count",
    Heart_Disease_Rate="mean"
).reset_index()

bmi_summary["Heart_Disease_Rate"] = bmi_summary["Heart_Disease_Rate"] * 100

bmi_summary
```

```python
# Step 20: Save BMI category summary

bmi_summary.to_csv("../tableau_dashboard_data/bmi_category_heart_disease_rate.csv", index=False)
```

```python
# Step 21: Confirm saved Tableau files

os.listdir("../tableau_dashboard_data")
```
